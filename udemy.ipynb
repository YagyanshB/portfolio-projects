```python
# Get the value counts of the 'subject' column
subject_counts = df['subject'].value_counts()

sns.barplot(x=subject_counts.index, y=subject_counts.values)
plt.xlabel('Subject')
plt.ylabel('Count')
plt.title('Distribution of Courses by Subject')
plt.xticks(rotation=45, ha='right')
plt.show()
```
