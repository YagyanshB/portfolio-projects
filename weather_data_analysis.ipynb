```python
# 4 Renaming the Weather column to Weather Condition

df.rename(columns = {'Weather' : 'Weather Condition'}, inplace = True)

# inplace command is used to make the change permanently
```

```python
df.head(2)
```

```python
# 6 - Calculating the Mean Visibility

visibility_mean = df['Visibility_km'].mean()

print("The mean visibility is", round(visibility_mean,2))
```

```python
# 7 - Calculating the Standard Deviation of Pressure

pressure_std = df['Press_kPa'].std()

print("The standard deviation of pressure is", round(pressure_std,2))
```

```python
# 8 - Calculating the variance of Relative Humidity

humidity_var = df['Rel Hum_%'].var()

print("The variance of relative humidity is", round(humidity_var,2))
```

```python
# 9 - Finding all instances when snow was recorded

df.head(2)


df[df['Weather Condition'] == 'Snow'].value_counts()
df[df['Weather Condition'].str.contains('Snow')]
```

```python
# 10 - Finding all instances when Wind Speed is above 24 and Visibility is 25

df[(df['Wind Speed_km/h']> 24) & (df['Visibility_km'] == 25)]
```

```python
# 11 - Mean value of each column against their weather condition

# Selecting only numeric columns for calculating mean
numeric_df = df.select_dtypes(include=np.number)

# Include 'Weather Condition' column in numeric_df
numeric_df = pd.concat([numeric_df, df['Weather Condition']], axis=1)

# Grouping by 'Weather Condition' and calculating the mean for numeric columns
mean_values = numeric_df.groupby('Weather Condition').mean()

mean_values
```

```python
# 12 - minimum and maximum value of each column

df.groupby('Weather Condition').agg(['min', 'max'])
df.groupby('Weather Condition').agg(['min', 'max']).reset_index()

df.groupby('Weather Condition').min()
df.groupby('Weather Condition').max()
```

```python
# 13 - show all records where weather conditions is fog

df[df['Weather Condition'].str.contains('Fog')]


df[df['Weather Condition'] == 'Fog']
```

```python
# 14 - finding instances when visibility is above 40 and weather is clear

df[(df['Visibility_km'] > 40) & (df['Weather Condition'] == 'Clear')]
```

```python
# 15 - Weather is clear and Relative Humidity is above 50 or Visibility is above 40

df[(df['Visibility_km'] > 40) | ((df['Weather Condition'] == 'Clear') & (df['Rel Hum_%'] > 50))]
```
